# 第2讲 凸集(Convex set)

## 2.1 仿射集

首先给出直线的数学定义  
>**<font color="blue"/>定义(直线)**   给定两个不同的点$x$和$y$，经过它们的直线定义为
$$ \theta x + (1-\theta) y$$
其中$\theta\in\mathbb{R}$，$x$和$y$均为$n$维向量

基于直线的定义，我们可以进一步给出仿射集的定义
>**<font color="blue"/>定义(仿射集)**   给定集合$C$内两个不同的点$x$和$y$，如果对于任意的$\theta\in\mathbb{R}$
$$ \theta x + (1-\theta) y\in C$$
成立，则称$C$为仿射集(affine set)  

我们可以从如下的角度来理解仿射集：  
在一个点集中任取两个不同的点$x$和$y$，如果经过它们的直线上的任意一点都属于这个点集，那么就说这个点集是一个仿射集。  
最常见的仿射集是线性方程租的解集$\{x|Ax=b\}$，这很容易证明:  
在解集中任取两点$x_1,x_2$，则  
$$Ax_1=b\\ Ax_2=b$$
接着考察
$$A\big(\theta x_1 + (1-\theta) x_2\big)=\theta A x_1+(1-\theta)A x_2=\theta b+(1-\theta) b=b$$
证毕  
实际上，这个结论反过来也是成立的，即任何仿射集都可以表示成一个线性方程租的解集。

## 2.2 凸集

在介绍凸集之前，先给出线段的数学定义
>**<font color="blue"/>定义(线段)**   给定两个不同的点$x$和$y$，经过它们的线段定义为
$$ \theta x + (1-\theta) y$$
其中$0\leq \theta\leq 1$，$x$和$y$均为$n$维向量

接着给出凸集的定义
>**<font color="blue"/>定义(凸集)**   给定集合$C$内两个不同的点$x$和$y$，如果对于任意的$0\leq\theta\leq 1$，下式都成立
$$ \theta x + (1-\theta) y\in C$$
则称$C$为凸集(convex set)，点集$\theta x + (1-\theta y)$称为x和y的凸组合。  

简单来说，凸集就是包含了经过这个集合中任意两个不同点间的线段的点集。

## 2.3 凸组合与凸包

>**<font color="blue"/>定义(凸组合)**   给定点集$C$内任意$K$个不同的点$x_1,x_2,...,x_K$，则
$$ x=\theta_1 x_1 + ...+\theta_K x_K$$
称为$x_1,x_2,...,x_K$的凸组合(convex combination),其中$0\leq \theta_k \leq 1，\sum_{k=1}^K \theta_k=1$

这里$C$不必是凸集。

>**<font color="blue"/>定义(凸包)**   给定点集$C$，它的凸包(convex hull) 定义为$C$内点的所有凸组合组成的点集
$$ conv C=\{x|x=\theta_1 x_1 + ...+\theta_K x_K\}$$
其中$0\leq \theta_k \leq 1，\sum_{k=1}^K \theta_k=1$

下面给出一段计算凸包的python代码示例：

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.spatial import ConvexHull
points = np.random.rand(30, 2)   # 在2-D空间生成30个随机点 
hull = ConvexHull(points)
# 绘制这些点
plt.plot(points[:,0], points[:,1], 'o')
# 绘制凸包多边形
for simplex in hull.simplices:
    plt.plot(points[simplex, 0], points[simplex, 1], 'k-')
plt.plot(points[hull.vertices,0], points[hull.vertices,1], 'r--', lw=2)
plt.show()
```

## 2.4 凸锥

>**<font color="blue"/>定义(锥组合)**   给定两个不同的点$x_1$和$x_2$，则
$$ x=\theta_1 x_1+\theta_2 x_2$$
称为点$x_1$和$x_2$的锥组合(conic combination)，其中$\theta_1,\theta_2 \geq 0$

接下来定义凸锥

>**<font color="blue"/>定义(凸锥)**   给定点集$C$中任意两个不同的点$x_1$和$x_2$，则
$$ \{x|x=\theta_1 x_1+\theta_2 x_2\}$$
称为$C$的凸锥(convex cone)，其中$\theta_1,\theta_2 \geq 0$

下图展示了凸锥的形状，可以看到凸锥由$x_1,x_2$定义的凸锥是由原点分别向$x_1$和$x_2$发出的射线包围成的区域组成的。

## 2.5 超平面和半空间

>**<font color="blue"/>定义(超平面)**  形如
$$ \{x|a^Tx=b\}$$
的点集称为超平面(hyperplane)，其中$a\neq 0$是超平面的法向量(normal vector)
 
简单来说超平面由空间中与法向量垂直的向量构成，当$b$发生变化时，超平面沿着法向量的方向平行移动。

<img src="https://ogu8vojlj.qnssl.com/%E9%80%89%E5%8C%BA_339.png" width=500/>

>**<font color="blue"/>定义(半空间)**  形如
$$ \{x|a^Tx\leq b\}$$
的点集称为半空间(halfspace)，其中$a\neq 0$

下图展示了半空间
<img src="https://ogu8vojlj.qnssl.com/%E9%80%89%E5%8C%BA_340.png" width=500/>


半空间可以理解为所有满足$a^Tx\leq b$的超平面构成的空间，容易证明，超平面既是仿射集也是凸集，而半空间仅是凸集。

## 2.6 超球和椭球

>**<font color="blue"/>定义(超球)**  满足
$$ \{x|\left\|x-x_0\right\|_2\leq r\}$$
的点集称为超球(Euclidean ball)，其中$x_0$为超球的球心，而$r$为超球的半径

超球可以认为是二位平面上的圆在多维空间上的推广。


>**<font color="blue"/>定义(椭球)**  满足
$$ \{x|(x-x_0)^TP^{-1}(x-x_0)\leq 1 \}$$
的点集称为椭球(ellipsoid)，其中$x_0$为超球的球心，$P$是对称半正定矩阵

矩阵$P$定义了椭球沿着$x_0$的各个方向延伸的长度，各个半轴的长度由$\sqrt{\lambda_i}$，其中$\lambda_i$是$P$的特征值。当$P=r^2I$时，椭球退化为超球，其中$I$是单位矩阵。  
椭球还可以有另一种表示：
$$\{x_0+Au|\left\|u\right\|_2\leq 1\}$$
其中$A$是一个非奇异的方阵  
下段代码演示了如何用python生成一个椭球：

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)

x = 10 * np.outer(np.cos(u), np.sin(v))
y = 10 * np.outer(np.sin(u), np.sin(v))
z = 10 * np.outer(np.ones(np.size(u)), np.cos(v))
ax.plot_surface(x, y, z, rstride=4, cstride=4, color='b')

plt.show()
```

## 2.7 标准球和标准锥

假设$\left\|\cdot\right\|$是任意的范数，我们依此定义：
>**<font color="blue"/>定义(标准球)**  满足
$$ \{x|\left\|x-x_0\right\|\leq r\}$$
的点集称为标准球(norm ball)，其中$x_0$为norm ball的球心，而$r$为norm ball的半径


>**<font color="blue"/>定义(标准锥)**  满足
$$ \{(x,t)|\left\|x\right\|\leq t\}$$
的点集称为标准锥(norm cone)，其中$t\in\mathbb{R}$为与$x$无关的实数。特别地，当选择欧氏距离作为范数时，我们得到二阶锥(second-order cone)，如下图所示。
